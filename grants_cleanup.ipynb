```python
import pandas as pd
import numpy as np
import json
import geonamescache
from geopy.geocoders import Nominatim
```

```python
not_states = ["MP", "PR", "AS", "UM", "VI", "GU"]
with open("tools/state_codes.json", 'r') as f:
    state_codes = json.load(f)
    print(state_codes)
    states_list = [key for key in state_codes if key not in not_states and state_codes[key] not in not_states]
```

```python
gl = Nominatim(user_agent="hacklytics23")
gc = geonamescache.GeonamesCache()
```

```python
pd.set_option('display.max_rows', None)
pd.options.mode.chained_assignment = None
```

```python
def city_to_county(city, state):
    print("C:" + city, state)
    cty = [c for c in gc.search_cities(city, case_sensitive=False) if c['admin1code'] == state][0]
    coords = cty["latitude"], cty["longitude"]
    result = gl.reverse(coords).address.split(", ")[-4]
    print(result)
    return result

def isState(state):
    return str(state).title() in state_codes

def decountify(s):
    if (s == "Carson City"):
        return
        
    return " ".join([i for i in s.split() if i.lower() not in ["county", "parish", "municipality", "borough", "town", "city"]])
```

```python
grants_data = "datasets/Emergency_Solutions_Grantee_(ESG)_Areas.csv"
```

```python
grants = pd.read_csv(grants_data)
grants = grants[["NAME", "STUSAB", "STATE", "TYPE", "CDBG_AMT", "ESG_AMT", "HOME_AMT", "HOPWA_AMT", "TOT_AMT"]]
grants = grants[grants["STUSAB"].isin(states_list)]
grants
```

```python
counties = grants[grants["TYPE"] == 61]
counties['NAME'] = counties['NAME'].apply(lambda x: x.title())
counties = counties.drop(columns=['STATE', 'TYPE']).rename(columns={'NAME':'county', 'STUSAB':'state'})
# counties
```
