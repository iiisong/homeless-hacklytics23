```python
import pandas as pd
import numpy as np
import json
```

```python
with open("tools/state_codes.json", 'r') as f:
    state_codes = json.load(f)
```

```python
pd.set_option('display.max_rows', None)
```

```python
age_data = "datasets/ACSST1Y2021.S0101-Data.csv"
```

```python
age = pd.read_csv(age_data)
age_temp = age
age_temp.columns = age_temp.iloc[0]
age_temp = age_temp.drop(0)
```

```python
age_temp['Percent 20-49'] = age_temp['Estimate!!Percent!!Total population!!AGE!!20 to 24 years'].astype(float) + age_temp['Estimate!!Percent!!Total population!!AGE!!25 to 29 years'].astype(float) + age_temp['Estimate!!Percent!!Total population!!AGE!!25 to 29 years'].astype(float) + age_temp['Estimate!!Percent!!Total population!!AGE!!30 to 34 years'].astype(float) + age_temp['Estimate!!Percent!!Total population!!AGE!!35 to 39 years'].astype(float) + age_temp['Estimate!!Percent!!Total population!!AGE!!40 to 44 years'].astype(float) + age_temp['Estimate!!Percent!!Total population!!AGE!!45 to 49 years'].astype(float)
age_temp['Percent Male 20-49'] = age_temp['Estimate!!Percent Male!!Total population!!AGE!!20 to 24 years'].astype(float) + age_temp['Estimate!!Percent Male!!Total population!!AGE!!25 to 29 years'].astype(float) + age_temp['Estimate!!Percent Male!!Total population!!AGE!!25 to 29 years'].astype(float) + age_temp['Estimate!!Percent Male!!Total population!!AGE!!30 to 34 years'].astype(float) + age_temp['Estimate!!Percent Male!!Total population!!AGE!!35 to 39 years'].astype(float) + age_temp['Estimate!!Percent Male!!Total population!!AGE!!40 to 44 years'].astype(float) + age_temp['Estimate!!Percent Male!!Total population!!AGE!!45 to 49 years'].astype(float)
age_temp = age_temp[['Geography', 'Geographic Area Name', 'Percent 20-49', 'Percent Male 20-49', 'Estimate!!Total!!Total population!!SUMMARY INDICATORS!!Median age (years)']]
age_temp.rename(columns={'Estimate!!Total!!Total population!!SUMMARY INDICATORS!!Median age (years)': 'Median Age'}, inplace=True)
```

```python
def getState(s):
    '''
    s: String containing 'County, State'
    return: String containing 'State Abbreviation'
    '''
    arr = s.split(', ')
    return state_codes[arr[1]]

def getCounty(s):
    '''
    s: String containing 'County, State'
    return: String containing 'County'
    '''
    arr = s.split(', ')
    return arr[0]
```

```python
age_temp2 = age_temp
age_temp2['County'] = age_temp2['Geographic Area Name'].apply(getCounty)
age_temp2['State'] = age_temp2['Geographic Area Name'].apply(getState)
age_temp2 = age_temp2.drop(columns=['Geographic Area Name', 'Geography'])
age_temp2 = age_temp2.rename(columns={'County':'county', 'State':'state'})
```
