```python
import pandas as pd
import numpy as np
```

```python
uer_data = "datasets/laucnty21.xlsx"
```

```python
pd.set_option('display.max_rows', None)
```

```python
uer = pd.read_excel(uer_data)
uer_temp = uer
uer_temp.columns = uer_temp.iloc[3]
uer_temp = uer_temp.drop([0,1,2,3,4])[["County Name/State Abbreviation", "Force", "(%)"]]
uer_temp[['county', 'state']] = uer_temp["County Name/State Abbreviation"].apply(lambda x: pd.Series(str(x).split(", ")))
uer_temp = uer_temp[~(uer_temp['state'].isin(['PR']))].dropna()
uer_temp = uer_temp[['state', 'county', 'Force', '(%)']].rename(columns={'Force':'force', '(%)': 'ue%'}).sort_values(['state', 'county'])
uer_temp.reset_index()
uer_temp.at[2, 'county'] = "Anchorage Municipality"
```

```python
uer = uer_temp
```

```python
uer.to_csv("./dataframes/uer")
```
