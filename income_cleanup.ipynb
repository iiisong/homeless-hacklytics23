```python
import pandas as pd
import numpy as np
import json
```

```python
not_states = ["MP", "PR", "AS", "UM", "VI", "GU"]
with open("tools/state_codes.json", 'r') as f:
    state_codes = json.load(f)
    states_list = [key for key in state_codes if key not in not_states and state_codes[key] not in not_states]
```

```python
with open("tools/state_codes.json", 'r') as f:
    state_codes = json.load(f)
```

```python
pd.set_option('display.max_rows', None)
pd.options.mode.chained_assignment = None
```

```python
inc_data = "datasets/lapi1122.xlsx"
```

```python
inc = pd.read_excel(inc_data)
inc.columns = ['county', '', '', 'inc','', '% change', '', '']
inc = inc.drop([0,1,2,3,4])
inc = inc[['county', 'inc', '% change']]
inc
```

```python
data = []
```

```python
state = ""
for index, row in inc.iterrows():
    if row['county'] in states_list:
        state = row['county']
        continue
    
#     print(state_codes[state])
    data.append([state_codes[state], row['county'], row['inc'], row['% change']])
```

```python
inc_data = pd.DataFrame(data, columns=['state', 'county', 'inc', '% change']).dropna()
```
